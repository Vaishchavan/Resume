```python
#data preprocessing
#helping function is convert
 #there is object
    #loop..loop will find dictionary
import ast    
def convert(obj):#here the genre is in string
    L=[]
    for i in ast.literal_eval(obj):#it is one func in ast which converts string into list for looping
        L.append(i['name'])
    return L    
     
```

```python
movies['genres']=movies['genres'].apply(convert)
```

```python
movies.head()#genres is sorted
```

```python
movies['keywords']=movies['keywords'].apply(convert)
```

```python
movies.head()#keywords is sorted
```

```python
#in cast keep only top 3 actors or dicionary
import ast    
def convert3(obj):#here the genre is in string
    L=[]
    counter=0
    for i in ast.literal_eval(obj):#it is one func in ast which converts string into list for looping
        if counter != 3:
            L.append(i['name'])
            counter+=1
        else:
            break
    return L 
#if loop for 3 times
     
```

```python
movies['cast']=movies['cast'].apply(convert3)
```

```python
movies.head()#cast is sorted
```

```python
#from crew only extract job=director
import ast    
def fetch_director(obj):#here the genre is in string
    L=[]
    for i in ast.literal_eval(obj):#it is one func in ast which converts string into list for looping
           if i['job']=='Director':
                L.append(i['name'])
                break
    return L    
     
```

```python
movies['crew']=movies['crew'].apply(fetch_director)
```

```python
from nltk.stem.porter import PorterStemmer
ps=PorterStemmer()
```

```python
def stem(text):
    y=[]#create a list
    
    for i in text.split(): #string is convertesd to list
        y.append(ps.stem(i))
        
    return " ".join(y)    
```

```python
new_df['tags']=new_df['tags'].apply(stem)
```

```python
new_df.head() #we hv 4806 vectors with 5000 max used words
```

```python
#calculate the distance of vector of one movie with the distance of vector of another movie
#recomend the movies that are closer
#gretaer the distance ...similarity is minimum
#will calculate the cosine distance..the angle between the each vector movie
```

```python
from sklearn.metrics.pairwise import cosine_similarity #silimarity is between 0 and 1 ..if it is 1 then siliarity id high
```

```python
cosine_similarity(vectors)
```

```python
similarity = cosine_similarity(vectors)
```

```python
list(enumerate(similarity[0])) # it consist of index of movie and similarity with it
```

```python
#apply sorting
sorted(list(enumerate(similarity[0])),reverse=True,key=lambda x:x[1])[1:6]# dont sort according to index...sort on the basis of second value
```

```python
#cretae a function recommend ...which in return will give five best movies 
def recommend(movie):
    movie_index = new_df[new_df['title']==movie].index[0]#find the index of movie
    distances=similarity[movie_index]#find the similarity
    #sort in the descending order
    #but indexing is lost in sort thats why enmumerate function is used
    movies_list= sorted(list(enumerate(distances)),reverse=True,key=lambda x:x[1])[1:6]

    for i in movies_list:
        print(new_df.iloc[i[0]].title)
    return
```

```python
recommend('Batman Begins')
```
